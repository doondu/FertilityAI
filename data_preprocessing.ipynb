```python
nan = {
    '기증 제공': 0,
    '본인 제공': 1,
    '알 수 없음': 2
}
jung = {
    '기증 제공': 0,
    '미할당': 1,
    '배우자 및 기증 제공': 2,
    '배우자 제공': 3
}

df['난자 출처'] = df['난자 출처'].map(nan)
df['정자 출처'] = df['정자 출처'].map(jung)
```

##### mutli-category -> one-hot

```python
df['특정 시술 유형'] = df['특정 시술 유형'].fillna('Unknown')
df.info()
```

```python
df['특정 시술 유형'].unique()
```

```python
df['특정 시술 유형 리스트'] = df['특정 시술 유형'].str.replace(':', '/').str.split('/')
df['특정 시술 유형 리스트'] = df['특정 시술 유형 리스트'].apply(lambda x: [s.strip() for s in x])

mlb = MultiLabelBinarizer()
one_hot = pd.DataFrame(mlb.fit_transform(df['특정 시술 유형 리스트']), columns=mlb.classes_)
df = pd.concat([df.drop(columns = ['특정 시술 유형 리스트']), one_hot], axis = 1)
df.drop(columns = ['특정 시술 유형'], inplace=True)

df.head()
```

```python
cols = ['AH', 'BLASTOCYST', 'FER', 'GIFT', 'Generic DI', 'ICI', 'ICSI', 'IUI', 'IVF', 'IVI', 'Unknown']
df[cols] = df[cols].astype(bool)
df.info()
```

```python
mlb = MultiLabelBinarizer()
one_hot = pd.DataFrame(mlb.fit_transform(df['배아 생성 주요 이유'].str.split(', ')), columns=mlb.classes_)
df = pd.concat([df, one_hot], axis = 1)
df.drop(columns = ['배아 생성 주요 이유'], inplace=True)
df.head()
```
